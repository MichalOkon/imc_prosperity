```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data/island-data-bottle-round-3/prices_round_3_day_0.csv', delimiter=';')
data = data.append(pd.read_csv('data/island-data-bottle-round-3/prices_round_3_day_1.csv', delimiter=';'))
data = data.append(pd.read_csv('data/island-data-bottle-round-3/prices_round_3_day_2.csv', delimiter=';'))

berry = data[data['product']=='BERRIES'].reset_index()
dolphs = data[data['product']=='DOLPHIN_SIGHTINGS'].reset_index()
divG = data[data['product']=='DIVING_GEAR'].reset_index()
plt.plot(divG['mid_price'])
```

```python
def get_std_and_mean(from_data, mean_window, std_window):
    stds = []
    means = []
    for time in dolphs.index:
        if time - mean_window < 0:
            stds.append(0)
            means.append(0)
        elif time - std_window < 0:
            ys = from_data['mid_price'][time - mean_window:time]
            means.append(np.mean(ys))
            stds.append(0)
        else:
            ys_mean = from_data['mid_price'][time - mean_window:time]
            ys_std = from_data['mid_price'][time - std_window:time]
            means.append(np.mean(ys_mean))
            stds.append(np.std(ys_std))
    return stds, means
```

```python
dolphs_stds, dolphs_means = get_std_and_mean(dolphs, mean_window=100, std_window=300)

## closer test:
    # 100, 200
```

```python
lo = 6000
hi = 7000
mult = 4
# plt.plot(dolphs['mid_price'][lo:hi])
plt.plot(dolphs['mid_price'][lo:hi])
plt.fill_between(dolphs.index[lo:hi], list(map(lambda aa,bb: aa + mult * bb, dolphs_means, dolphs_stds))[lo:hi], list(map(lambda aa,bb: aa - mult * bb, dolphs_means, dolphs_stds))[lo:hi], color='orange')

for i in range(lo, hi):
    price = dolphs['mid_price'][i]
    lower_bound = dolphs_means[i] - mult * dolphs_stds[i]
    upper_bound = dolphs_means[i] + mult * dolphs_stds[i]
    if price < lower_bound:
        print("LOWER", i)
    if price > upper_bound:
        print("HIGHER", i)
```

```python
lo = 500
hi = 30000

in_peak_sold = False
in_peak_bought = False

cash = 0
pos = 0
last_price = 0
for i in range(lo, hi):
    divG_price = divG['mid_price'][i]
    if not in_peak_sold and not in_peak_bought:
        dolph_price = dolphs['mid_price'][i]
        lower_bound_dolphs = dolphs_means[i] - 4 * dolphs_stds[i]
        upper_bound_dolphs = dolphs_means[i] + 4 * dolphs_stds[i]
        if all(x < lower_bound_dolphs for x in dolphs['mid_price'][i-2:i]):
            assert not in_peak_bought
            in_peak_sold = True
            print("SELL", i)
            cash += divG_price
            pos -= 1
        elif all(x > upper_bound_dolphs for x in dolphs['mid_price'][i-2:i]):
            assert not in_peak_sold
            in_peak_bought = True
            print("BUY", i)
            cash -= divG_price
            pos += 1
    else:
        lower_bound_divG = divG_means[i] - 2 * divG_stds[i]
        upper_bound_divG = divG_means[i] + 2 * divG_stds[i]
        if in_peak_sold and all(x > upper_bound_divG for x in divG['mid_price'][i-10:i]):
            assert not in_peak_bought
            in_peak_sold = False
            print("BUY BACK", i)
            cash -= divG_price
            pos += 1
        elif in_peak_bought and all(x < lower_bound_divG for x in divG['mid_price'][i-10:i]):
            assert not in_peak_sold
            in_peak_bought = False
            print("SELL BACK", i)
            cash += divG_price
            pos -= 1
    last_price = divG_price
# print(cash, pos)
print("shomeny with 1 lot per order:", cash + pos * last_price)
```
