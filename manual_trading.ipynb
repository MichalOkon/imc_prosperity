```python
import numpy as np
```

```python
exchange = np.array([
    [1,  0.5, 1.45, 0.75],
    [1.95,  1, 3.1, 1.49],
    [0.67, 0.31, 1, 0.48],
    [1.34, 0.64, 1.98, 1]
])
```

```python
commodities = ['Pizza Slice', 'Wasabi Root', 'Snowball', 'Shells']
```

```python
def perform_exchange(seq):
    seq = [3, *seq, 3]
    s = 1
    for i in range(len(seq) - 1):
        s *= exchange[seq[i]][seq[i + 1]]
    return s
```

```python
perform_exchange([0, 0])
```

```python
results = []
```

```python
for i in range(len(commodities)):
    curr_list_i = [i]
    results.append((curr_list_i, perform_exchange(curr_list_i)))
    for j in range(len(commodities)):
        
        if curr_list_i[-1] == j:
            continue
        curr_list_j = [*curr_list_i, j]
        results.append((curr_list_j, perform_exchange(curr_list_j)))
        for k in range(len(commodities)):
            
            if curr_list_j[-1] == k:
                continue
            curr_list_k = [*curr_list_j, k]
            results.append((curr_list_k, perform_exchange(curr_list_k)))
            for l in range(len(commodities)):
                
                if curr_list_k[-1] == l:
                    continue
                curr_list_l = [*curr_list_k, l]
                results.append((curr_list_l, perform_exchange(curr_list_l)))
    
```

```python
sorted(results, key=lambda x: -x[1])
```
